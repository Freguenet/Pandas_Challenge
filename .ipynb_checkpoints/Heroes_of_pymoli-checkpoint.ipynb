```python
import pandas as pd
```

```python
file = ("purchase_data.csv")
```

```python
pymoli_df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
pymoli_df.head()
```

## Player Count

```python
total_players = pymoli_df["SN"].nunique()
player_count = {"": [""], 
                "Total Players": [total_players]}
df_player_count = pd.DataFrame(player_count)
df_player_count
```

## Purchasing Analysis (Total)

```python
#purchasing annalysis
num_unique = pymoli_df["Item Name"].nunique()
ave_price = (pymoli_df["Price"].mean())
num_purchase = pymoli_df["Purchase ID"].count()
total_purchase = pymoli_df["Price"].sum()
purch_annalysis = {"": [""], 
                "Unique Items": [num_unique],
                  "Average Price": [ave_price],
                   "Number of Purchases": [num_purchase],
                  "Total Purchase": [total_purchase]}
df_purchase_annalysis = pd.DataFrame(purch_annalysis)
df_purchase_annalysis["Average Price"] = df_purchase_annalysis["Average Price"].astype(float).map("${:,.3}".format)
df_purchase_annalysis["Total Purchase"] = df_purchase_annalysis["Total Purchase"].astype(float).map("${:,.6}".format)
df_purchase_annalysis
```

## Gender Demographics

```python
#gender demographic
grouped_gender = pymoli_df.groupby("Gender")
grouped_gender_df = pd.DataFrame(grouped_gender["SN"].nunique())

grouped_gender_purch_tot_df = pd.DataFrame(grouped_gender["Purchase ID"].count())
grouped_gender_ave_price_df = (pd.DataFrame(grouped_gender["Price"].mean())).round(2)

grouped_gender_tot_purch_df = pd.DataFrame(grouped_gender["Price"].sum())
total_purch_df = (pd.DataFrame(grouped_gender["Price"].sum())).round(2)
ave_total_purch_pers = (pd.DataFrame(grouped_gender["Price"].sum()/grouped_gender["SN"].nunique())).round(2)

perc_gender = (grouped_gender["Purchase ID"].nunique()/num_purchase)*100
perc_gender_df = (pd.DataFrame(perc_gender)).round(2)

merged_gender_df = pd.merge(grouped_gender_df, perc_gender_df, on=["Gender"])
merged_gender_df.columns = ["Count of Players", "Percentage of Players"]
merged_gender_df
```

## Purchasing Analysis (Gender)

```python
purch_annalysis_gender = pd.merge(grouped_gender_purch_tot_df, grouped_gender_ave_price_df, on="Gender").merge(total_purch_df, on="Gender").merge(ave_total_purch_pers, on="Gender")
purch_annalysis_gender_df = pd.DataFrame(purch_annalysis_gender)
purch_annalysis_gender_df.columns = ["Number of Purchase", "Avg Purchase Price", "Total Purchase Value", "Avg Purchase per Person"]
purch_annalysis_gender_df["Avg Purchase Price"] = purch_annalysis_gender_df["Avg Purchase Price"].astype(float).map("${:,.4}".format)
purch_annalysis_gender_df["Total Purchase Value"] = purch_annalysis_gender_df["Total Purchase Value"].astype(float).map("${:,.6}".format)
purch_annalysis_gender_df["Avg Purchase per Person"] = purch_annalysis_gender_df["Avg Purchase per Person"].astype(float).map("${:,.4}".format)
purch_annalysis_gender_df
```

##  Age Demographics

```python
#Demo
print(pymoli_df["Age"].max())
print(pymoli_df["Age"].min())
```

```python
bins = [0,9, 14, 19, 24, 29, 34, 39, 45]

group_labels = ["< 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pd.cut(pymoli_df["Age"], bins, labels=group_labels).head()
pymoli_df["Age Group"] = pd.cut(pymoli_df["Age"], bins, labels=group_labels)
pymoli_df.head()
```

```python
player_age = pymoli_df.groupby("Age Group")

player_age_df = pd.DataFrame(player_age["SN"].nunique())
player_age_df
perc_age = (player_age["SN"].nunique()/total_players)*100
perc_age_df = pd.DataFrame(perc_age)
perc_age_df
merged_age_perc = pd.merge(player_age_df, perc_age_df, on="Age Group")
merged_age_perc.columns = ["Number of Players", "Percentage of Players"]
merged_age_perc["Percentage of Players"] = merged_age_perc["Percentage of Players"].astype(float).map("{:,.3}%".format)
merged_age_perc
```

##  Purchasing Analysis (Age)

```python
#Purch annalysis by age
player_age_ave_purch_df = pd.DataFrame(player_age["Price"].mean())
player_age_purch_df = pd.DataFrame(player_age["Purchase ID"].count())
#player_age_ave_purch_df
player_age_sum_purch_df = pd.DataFrame(player_age["Price"].sum())
#player_age_sum_purch_df
player_age_ave_person_purch = player_age["Price"].sum()/player_age["SN"].nunique()
player_age_ave_person_purch_df = (pd.DataFrame(player_age_ave_person_purch))

```

```python
purch_annalysis_age = pd.merge(player_age_purch_df, player_age_ave_purch_df, on="Age Group").merge(player_age_sum_purch_df, on="Age Group").merge(player_age_ave_person_purch_df, on="Age Group")
purch_annalysis_age_df = pd.DataFrame(purch_annalysis_age)
purch_annalysis_age_df.columns = ["Number of Purchase", "Avg Purchase", "Total Purchase", "Avg Purchase by Player"]
purch_annalysis_age_df["Avg Purchase"] = purch_annalysis_age_df["Avg Purchase"].astype(float).map("${:,.3}".format)
purch_annalysis_age_df["Total Purchase"] = purch_annalysis_age_df["Total Purchase"].astype(float).map("${:,.6}".format)
purch_annalysis_age_df["Avg Purchase by Player"] = purch_annalysis_age_df["Avg Purchase by Player"].astype(float).map("${:,.3}".format)
purch_annalysis_age_df
```

##  Top Spenders

```python
#top spender
grouped_total_players = pymoli_df.groupby("SN")

purchase_count_df = pd.DataFrame(grouped_total_players["Price"].count())
ave_top_purch_price_df = (pd.DataFrame(grouped_total_players["Price"].mean()))
total_top_purch_price_df = pd.DataFrame(grouped_total_players["Price"].sum())
tops_spender_annalysis = pd.merge(purchase_count_df, ave_top_purch_price_df, on="SN").merge(total_top_purch_price_df, on="SN")
tops_spender_annalysis_df = pd.DataFrame(tops_spender_annalysis)

top_spender_df = pd.DataFrame(tops_spender_annalysis_df.sort_values(by=["Price"],ascending=False))
top_spender_df.columns = ["Total Purchase", "Avg Purchase Price", "Total Purchase Value"]
top_spender_df["Avg Purchase Price"] = top_spender_df["Avg Purchase Price"].astype(float).map("${:,.3}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].astype(float).map("${:,.4}".format)
top_spender_df[:5]


```

## Most Popular Items

```python
grouped_items = pymoli_df.groupby(["Item ID", "Item Name"])
grouped_items_count_df = pd.DataFrame(grouped_items["Purchase ID"].count())
grouped_items_count_df
grouped_items_tot_value_df = pd.DataFrame(grouped_items["Price"].sum())
grouped_items_price = grouped_items["Price"].sum()/grouped_items["Purchase ID"].count() 
grouped_items_price_df = pd.DataFrame(grouped_items_price)
merged_items_count_annalysis = pd.merge(grouped_items_count_df, grouped_items_price_df, on=["Item ID", "Item Name"]).merge(grouped_items_tot_value_df, on=["Item ID", "Item Name"])
merged_items_count_annalysis_df = pd.DataFrame(merged_items_count_annalysis)
merged_items_count_annalysis_df = pd.DataFrame(merged_items_count_annalysis_df.sort_values(by=["Purchase ID"],ascending=False))
merged_items_count_annalysis_df.columns = ["Purchase Count", "Item Price", "Total Purchase Value"]
merged_items_count_annalysis_df["Item Price"] = merged_items_count_annalysis_df["Item Price"].astype(float).map("${:,.3}".format)
merged_items_count_annalysis_df["Total Purchase Value"] = merged_items_count_annalysis_df["Total Purchase Value"].astype(float).map("${:,.4}".format)
merged_items_count_annalysis_df[:5]
```

##  Most Profitable Items

```python

merged_items_profit_annalysis_df = pd.DataFrame(merged_items_count_annalysis)
merged_items_profit_annalysis_df.columns = ["Purchase Count", "Item Price", "Total Purchase Value"]
merged_items_profit_annalysis_df = pd.DataFrame(merged_items_profit_annalysis_df.sort_values(by=["Total Purchase Value"],ascending=False))
merged_items_profit_annalysis_df["Item Price"] = merged_items_profit_annalysis_df["Item Price"].astype(float).map("${:,.3}".format)
merged_items_profit_annalysis_df["Total Purchase Value"] = merged_items_profit_annalysis_df["Total Purchase Value"].astype(float).map("${:,.4}".format)
merged_items_profit_annalysis_df[:5]
```
